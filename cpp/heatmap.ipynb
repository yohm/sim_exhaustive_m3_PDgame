```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
# a = np.loadtxt('gtft.dat', delimiter=' ', dtype='int64')
# a = np.loadtxt('tft_atft.dat', delimiter=' ', dtype='int64')
# a = np.loadtxt('alld.dat', delimiter=' ', dtype='int64')
# a = np.loadtxt('capri.dat', delimiter=' ', dtype='int64')
# a

plt.rcParams['font.family'] ='sans-serif'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.width'] = 1.0
plt.rcParams['ytick.major.width'] = 1.0
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['font.size'] = 22
plt.rcParams['axes.linewidth'] = 1.2
```

```python
def make_fig(a, figname):
    c = 'gray'
    lw = 0.5
    x = np.linspace(0,1,100)
    y = np.linspace(4,1,100)
    plt.plot(x,y,color=c, linewidth=lw)
    x = np.linspace(1,4,100)
    y = np.linspace(1,0,100)
    plt.plot(x,y,color=c, linewidth=lw)
    x = np.linspace(0,3,100)
    y = np.linspace(4,3,100)
    plt.plot(x,y,color=c, linewidth=lw)
    x = np.linspace(3,4,100)
    y = np.linspace(3,0,100)
    plt.plot(x,y,color=c, linewidth=lw)
    x = np.linspace(1,3,100)
    y = np.linspace(1,3,100)
    plt.plot(x,y,color=c, linestyle='dotted', linewidth=0.5)
    plt.xlim(0,4)
    plt.ylim(0,4)
    plt.xticks([0,1,2,3,4])
    plt.yticks([0,1,2,3,4])
    plt.imshow(a, cmap='Blues', interpolation='antialiased', extent=(0,4,0,4), norm=colors.LogNorm())
    plt.savefig(figname)
    
```

```python
a = np.loadtxt('capri.dat', delimiter=' ', dtype='int64')
make_fig(a, 'capri.pdf')
```

```python
a = np.loadtxt('gtft.dat', delimiter=' ', dtype='int64')
make_fig(a, 'gtft.pdf')
```

```python
a = np.loadtxt('tft_atft.dat', delimiter=' ', dtype='int64')
make_fig(a, 'tft_atft.pdf')
```
